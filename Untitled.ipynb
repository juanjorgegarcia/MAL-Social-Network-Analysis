# ANIME

#### Topic Modeling

```python
g_animetopic = fm.load('animetopic.gml')
g_animetopic.set_all_nodes('labpos', 'hover')
# g_animetopic.move('bipartite', [n for n in g_animetopic.nodes if g_animetopic.nodes[n]['type'] == 'anime'])
# g_animetopic.draw()
```

```python
g_animetopic.set_nodecol('tipo', 'type')
g_animetopic.set_nodecol('nota', 'score')
g_animetopic.set_nodecol('titulo', 'title')
g_animetopic.set_nodecol('topico', 'topic')
g_animetopic.set_nodecol('corr', 'corr')
h, a = nx.hits(g_animetopic)
g_animetopic.set_nodecol('hub', h)
g_animetopic.set_nodecol('authority', a)
g_animetopic.nodeframe
```

```python
score_filter = g_animetopic.nodeframe.nota != 0.0
g_animetopic_filter = g_animetopic.nodeframe[score_filter]
g_animetopic_filter
```

```python
# df = g_animetopic_filter.groupby('titulo')['corr'].apply(lambda group_series:  group_series.tolist() ).reset_index()
df = g_animetopic_filter[g_animetopic_filter["corr"] > 0]
# df['test']= list(zip(df["corr"], df["topico"]))
df = df.groupby(['titulo', 'nota'])['topico'].apply(lambda group_series:  group_series.tolist() ).reset_index()
x = []
# modaDF = pd.DataFrame([{"anime1": '', 'anime2':'', 'sharedTopics': 0, "numberOfSharedTopics":0}])
# modaDF = pd.DataFrame(columns=['anime1', 'anime2', 'sharedTopics', "numberOfSharedTopics"])
```

```python
df2= df.head(2000)
df2
```

```python
for i in df2.iterrows():
    print(i[1][2])
```

```python
x = []
for i in df2.iterrows():
    anime1 = i[1][0]
    nota1 = i[1][1]
    lista1 = i[1][2]
    id1 = i[0]
    for j in df2.iterrows():
        id2 = j[0]
        if id1 < id2:
            anime2 = j[1][0]
            nota2 = j[1][1]
            lista2 = j[1][2]
            a = ((set(lista1).intersection(lista2)))
            data = {"anime1": anime1, "nota1": nota1, 'anime2':anime2, "nota2": nota2, 'sharedTopics':list(a), "numberOfSharedTopics":len(a)}
#             print(data)
            x.append(data)
modaDF = pd.DataFrame(x)
```

```python
modaDF.to_csv('modaDF2.csv', index=None)
```

```python
modaDF2 = pd.read_csv('modaDF2.csv')
```

```python
modaDF2
```

```python
moda2 = modaDF2.loc[:,"numberOfSharedTopics"].mode()
```

```python
moda2 = int(moda2)
moda2
```

```python
modaDF2.size / 4
```

```python
f = open('finaltopic.gml','w+', encoding="ascii")
f.write('graph [\n  directed 0\n')
animes = []

id = 0
for i in modaDF2.iterrows():
#     print(i[1])
#     adv
#     print(i[1][2])
#     print(i['numberOfSharedTopics'])
    if i[1][5] >= moda2:
        anime1 = i[1][0]
        nota1 = i[1][1]
        if anime1 not in animes:
            animes.append(anime1)
            f.write('    node [ id %d title "%s" score %d ]\n' % (id, anime1, nota1))
            id1 = id
            id += 1
            
            
        anime2 = i[1][2]
        nota2 = i[1][3]
        if anime2 not in animes:
            animes.append(anime2)
            f.write('    node [ id %d title "%s" score %d ]\n' % (id, anime2, nota2))
            f.write('  edge [ source %s target %s ]\n' % (id1, id))
            id += 1
              

f.write(' ]')
f.close()
```

```python
g_finaltopic = fm.load('finaltopic.gml')
g_finaltopic.set_all_nodes('labpos', 'hover')
g_finaltopic.move('kamada_kawai')
# g_finaltopic.draw()
```

```python
centrality = nx.eigenvector_centrality(g_finaltopic)

g_finaltopic.set_nodecol('nome', 'label')
g_finaltopic.set_nodecol('degree', g_finaltopic.degree)
g_finaltopic.set_nodecol('eigenvector', centrality)

df2 = g_finaltopic.nodeframe
# df2['authority'] = df['authority']
df2.sort_values('eigenvector', ascending=False)
```
