```python
import networkx as nx
import freeman as fm
import pandas as pd
```

# ANIME

#### Genre

```python
g_animegenre = fm.load('animegenre.gml')
g_animegenre.set_all_nodes('labpos', 'hover')
g_animegenre.move('bipartite', [n for n in g_animegenre.nodes if g_animegenre.nodes[n]['type'] == 'anime'])
# g_animegenre.draw()
```

```python
g_animegenre.set_nodecol('tipo', 'type')
h, a = nx.hits(g_animegenre)
g_animegenre.set_nodecol('hub', h)
g_animegenre.set_nodecol('authority', a)
g_animegenre.nodeframe
```

#### Topic Modeling

```python
g_animetopic = fm.load('animetopic.gml')
g_animetopic.set_all_nodes('labpos', 'hover')
g_animetopic.move('bipartite', [n for n in g_animetopic.nodes if g_animetopic.nodes[n]['type'] == 'anime'])
# g_animetopic.draw()
```

```python
g_animetopic.set_nodecol('tipo', 'type')
g_animetopic.set_nodecol('nota', 'score')
g_animetopic.set_nodecol('titulo', 'title')
g_animetopic.set_nodecol('topico', 'topic')
g_animetopic.set_nodecol('corr', 'corr')
h, a = nx.hits(g_animetopic)
g_animetopic.set_nodecol('hub', h)
g_animetopic.set_nodecol('authority', a)
g_animetopic.nodeframe
```

```python
score_filter = g_animetopic.nodeframe.nota != 0.0
g_animetopic_filter = g_animetopic.nodeframe[score_filter]
g_animetopic_filter
```

```python
g_animetopic_filter


# df = g_animetopic_filter.groupby('titulo')['corr'].apply(lambda group_series:  group_series.tolist() ).reset_index()
df = g_animetopic_filter[g_animetopic_filter["corr"] > 0]
# df['test']= list(zip(df["corr"], df["topico"]))
df = df.groupby('titulo')['topico'].apply(lambda group_series:  group_series.tolist() ).reset_index()
x = []
modaDF = pd.DataFrame()

for i in df.iterrows():
    anime1 = i[1][0]
    lista1 = i[1][1]
    id1 = i[0]
    for j in df.iterrows():
        anime2 = j[1][0]
        lista2 = j[1][1]
        id2 = j[0]
        if id1 > id2:
            a = ((set(lista1).intersection(lista2)))
            data = pd.DataFrame({"anime1": anime1, 'anime2':anime2, 'sharedTopics':a, "numberOfSharedTopics":len(a)})
            modaDF.append(data)
    
    
    
```
